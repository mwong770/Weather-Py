# WeatherPy

*This analysis uses weather data from 500+ random cities to visualize the weather at varying distances from the equator.*

### Analysis

* The weather was generally warmest, with a maximum temperature of approximately 85 degrees F, at the equator (0 degrees latitute) with the exception of 25 of the 618 cities analyzed, which had maximum tempertures up to 113 degrees F. 18 of the 25 cities with maximum temperatures greater than the equator were located between -2.5 degrees latitude and -35 degrees latitude.
* Temperatures generally decreased as the distance increased from the equator. The temperatures in the northern hemisphere (positive degrees latitude) decreased much more than temperatures in the southern hemisphere (negative degrees latitude). The lowest maximum temperture in the southern hemisphere was 34.67 degrees F, while the maximum temperature in the northern hemisphere was -25.61 degrees F. This may be due to cities in the southern hemisphere being closer to the equator than cities in the northern hemisphere. 
* Based on the scatter plot, there does not appear to be a strong relationship between degrees latitude and humidity. The mean humidity is slightly higher in the southern hemisphere (80%) than the northern hemisphere (74%). 
* Based on the scatter plot, there does not appear to be a strong relationship between degrees latitude and cloudiness. However, there were strong bands of cities with approximately 0%, 75%, and 90% cloudiness. 
* Based on the scatter plot, there does not appear to be a strong relationship between degrees latitude and wind speed. The mean wind speed was similar in the southern hemisphere (7.7 mph) and the northern hemisphere (8.2 mph). While the vast majority of cities had wind speeds below 25 mph, 9 of the 618 cities analyzed had wind speeds above 25 mph. All but one of the 9 cities were located between 40 degrees latitude and 66 degrees latitude.

```python
# Dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import os
import json

from config import api_key

from citipy import citipy
```

## Generate Cities List

```python
cities = []

# 1500 random lat and lng combination
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)

# lat, lng combination per tuple
lat_lngs = zip(lats, lngs)

# for each lat, lng combination
for lat_lng in lat_lngs:

    # finds nearest city 
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # stores city if unique
    if city not in cities:
        cities.append(city)

# confirm min of 500 cities
len(cities)
```

## Perform API Calls

```python
df = pd.DataFrame({
    'City': '',
    'Cloudiness': '',
    'Country': '',
    'Date': '',
    'Humidity': '',
    'Lat': '',
    'Lng': '',
    'Max Temp': '',
    'Wind Speed': ''
}, index=[0])

print(df.shape)
df.head()
```

```python
query_url = 'http://api.openweathermap.org/data/2.5/weather?'

params = {
    'appid': api_key,
    'units': 'Imperial',
    'q': city
}

df_idx_ct = 0
```

```python
print('Beginning Data Retrieval')
print('-------------------------')

for index, city in enumerate(cities):
    params['q'] = city
    try:
        
        print(' ')
        print(f'Processing Record {index} | {city}')
        
        response = requests.get(query_url, params=params).json()
        
        # populates df row
        df.loc[df_idx_ct, 'City'] = response['name']
        df.loc[df_idx_ct, 'Cloudiness'] = response['clouds']['all']
        df.loc[df_idx_ct, 'Country'] = response['sys']['country']
        df.loc[df_idx_ct, 'Date'] = response['dt']
        df.loc[df_idx_ct, 'Humidity'] = response['main']['humidity']
        df.loc[df_idx_ct, 'Lat'] = response['coord']['lat']
        df.loc[df_idx_ct, 'Lng'] = response['coord']['lon']
        df.loc[df_idx_ct, 'Max Temp'] = response['main']['temp_max']
        df.loc[df_idx_ct, 'Wind Speed'] = response['wind']['speed']
        
        df_idx_ct += 1
        
    except:
        print('City not found. Skipping...')
        pass
            
print('\n-------------------------')
print('Data Retrieval Complete')
print('-------------------------')
```

## Plotting the Data

```python
def build_lat_scatter(y_axis, column, y_label):
    
    plt.scatter(df['Lat'], y_axis, marker='o', 
                edgecolor='black', alpha=0.8)

    # %x returns local date representation
    plt.title(f'City Latitude vs. {column} ({time.strftime("%x")})', fontsize=14)
    plt.xlabel('Latitude', fontsize=12)
    plt.ylabel(y_label, fontsize=12)
    plt.grid()
    plt.show()
```
